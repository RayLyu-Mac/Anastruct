```python
#this python script will be used as the structre analysis for polygen support 
#input will be 
# 1. sides of the polygon
# 2. what kind of support you want to put on? 1: mid point connection, 2: side point connection, 3: (else) combined tyoe
# 3. then input the number of truss you want to put in the sturcture (the code will ask you to do multiple input according to your requirment)
  # 1. for mid point connection the first chosen point will be the mid point of the structure, input will be the anonther point that you want to connect
  # 2. for the side point connection: (can be improved): you need to enter separatly 2 points in the side you want to connect
# 4. input the what type of the load you want to act on the system:  details seen in the cell #5
#5. input the location/amount of the force into the /system 
from anastruct import SystemElements
import matplotlib.pyplot as plt
import numpy as np
import math 
ss=SystemElements()
```

```python
#this function will return the support type with the connection between side point 
def stretch(q,location):
    '''
    #this function will return the support type with the connection between side point 
    will enter the fixed point twice to comfirm the connection 
    '''
    number=int(input('put number of truss here: '))
    i=0
    while i!=number:
        fixed1=input('put your fixed point 1 here:')
        fixed2=input('put your fixed point 2 here:')
        if int(fixed1)>q or int(fixed2)>q :
            print('index out of range!! please enter a number which is within the number of sides of the polygen')
            i+=0
        else:
            x1=location[int(fixed1)-2][0]
            x2=location[int(fixed2)-2][0]
            y1=location[int(fixed1)-2][1]
            y2=location[int(fixed2)-2][1]
            ss.add_element([[x1,y1],[x2,y2]])
            i=i+1
    ss.add_support_fixed(node_id=1)
    ss.add_support_fixed(node_id=2)
    load(q)
    ss.solve()
    %time x=ss.show_structure(values_only=True)
    %time ss.show_structure()
    %time ss.show_results()
    ss.show_results(figsize=(15,15))
```

```python
#this function will return a combined structure of both mid point connection and side pint connection 
def combine(q,location):
    '''
    #this function will return a combined structure of both mid point connection and side pint connection 
    '''
    p=0.25*q+1
    xs=x+math.cos(0*np.pi/q*2)
    xmid=(xs+x)/2
    if q%2==0:
        if q%4==0:
            p=0.25*q-1
            p1=p+1
            ym=location[int(p)][1]
            ym1=location[int(p1)][1],location[int(p1)][1]
            ymid=(ym+ym1)/2
        else:
            p=q/4-0.5
            ymid=location[int(p)][1]
    else:
        p=q/2-0.5
        ymid=location[int(p)][1]/2
    number1=int(input('put number of truss here for mid connection: '))
    i=0
    while i!=number1:
        fixed1=input('put your fixed point here:')
        if int(fixed1)>q:
            print('index out of range!! please enter a number which is within the number of sides of the polygen')
            i+=0
        else:
            ss.add_element(location=[[xmid,ymid],[location[int(fixed1)-2][0],location[int(fixed1)-2][1]]])
            i=i+1
    number2=int(input('put number of truss here for stretch connection: '))
    k=0
    while k!=number2:
        fixed1=input('put your fixed point 1 here:')
        fixed2=input('put your fixed point 2 here:')
        if int(fixed1)>q or int(fixed2)>q :
            print('index out of range!! please enter a number which is within the number of sides of the polygen')
            k+=0
        else:
            ss.add_element(location=[[xmid,ymid],[location[int(fixed1)-2][0],location[int(fixed1)-2][1]]])
            x1=location[int(fixed1)-2][0]
            x2=location[int(fixed2)-2][0]
            y1=location[int(fixed1)-2][1]
            y2=location[int(fixed2)-2][1]
            ss.add_element([[x1,y1],[x2,y2]])
            k+=1
    ss.add_support_fixed(node_id=1)
    ss.add_support_fixed(node_id=2)
    load(q)
    ss.solve()
    %time ss.show_results()
    %time ss.show_structure()  
    
```

```python
#this function will return the type of loads: 
#morespecifically: point loads, moment loads, and q loads
# input will be the location, amount 
def load (q):
    '''
    #this function will return the type of loads: 
#morespecifically: point loads, moment loads, and q loads
# input will be the location, amount 
if the input is none of them then will output a typo
'''
    types=str(input('put what kind of load you want to act on the strucutre here: '))
    if types=='point':
        numbers_of_load=int(input('put number of load you want to act on: '))
        o=0
        while o!=numbers_of_load:
            location=int(input('put the location of load you want to act on(integer only): '))
            if location>q:
                print ('not exceed index!!!')
                o+=0
            else:
                amountx=int(input('put amount of force in x direction:'))
                amounty=int(input('put amount of force in y direction:'))
                ss.point_load(node_id=location,Fx=amountx,Fy=amounty)
                o+=1
    elif types=='moment':
        numbers_of_load=int(input('put number of load you want to act on: '))
        w=0
        while w!=numbers_of_load:
            location=int(input('put the location of load you want to act on(integer only): '))
            if location>q:
                print ('not exceed index!!!')
                w+=0
            else:
                amounty=int(input('put amount of force in y direction:'))
                ss.moment_load(node_id=location,Ty=amounty)
                w+=1
    elif types=='qload':
        numbers_of_load=int(input('put number of load you want to act on: '))
        y=0
        while y!=numbers_of_load:
            if location>q:
                print ('not exceed index!!!')
                y+=0
            else:
                amount=int(input('put amount of force:'))
                location=int(input('put the location of load you want to act on(integer only): '))
                ss.q_load(q=amount, element_id=location, direction='element')
                y+=1            
    else:
        if q%2==0:
            ss.point_load(node_id=int(q/2),Fy=-100)
        else:
            ss.point_load(node_id=int((q+1)/2+1),Fy=-100)
```

```python
ss=SystemElements()
'''
   this function is the main function
   q represent the number sides of the polygen (enter)
   then it will ask the to enter the type of the support
'''
[x,y]=[0,0]
q=int(input('please enter the sides of the polygon: '))
ss=SystemElements(figsize=(40,40))
location=[]
for i in range(0,q):
    [x,y]=[x+math.cos(i*np.pi/q*2),y+math.sin(i*np.pi/q*2)]
    location.append([x,y]) #location here will record the all coordinate of the x,y points for (the side)
    ss.add_truss_element([x,y]) 
choice=str(input('put you choice here,"support type" rather mid or stretch or combined'))
if choice=='mid':
    findmid(q,location)
elif choice=='stretch':
    stretch(q,location)
elif choice=='combine':
    combine(q,location)
else:
    load(q)
    ss.add_support_fixed(node_id=1)
    ss.add_support_fixed(node_id=2)
    ss.show_structure()
```
